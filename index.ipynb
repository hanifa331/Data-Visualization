#checking for missing values and filling

```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values)
```

```python
df.describe()
```

```python
missing_df=df.isnull().sum()
print(type(missing_df))
print(missing_df)
```

#dorting the missing values and convertin them into percentage

```python
sorted_df = missing_df.sort_values(ascending=False)
total_rows = df.shape[0]
missing_df_percentage = (missing_df / total_rows) * 100
missing_df_percentage = missing_df_percentage.sort_values(ascending=False).round(0)
```

#Identyfying the columns with more than 20$

```python
dropped_columns = missing_df_percentage > 20.0
```

#printing the columns with missing values

```python
print("\nMissing Values Percentage > 20:\n", missing_df_percentage[dropped_columns])
print("\nMissing Values Count:\n", sorted_df)
```

cleaned data

```python
df_cleaned = df.drop(columns=missing_df_percentage[missing_df_percentage > 20].index)
print("\nDataframe after dropping columns with >20% missing values:\n", df_cleaned.head())
```

# Accessing the columns and replacing the numerical columns with '0' and the categorical column with 'unknown'

```python
numerical_columns = df_cleaned.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
numerical_colums= ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries',
       'Total.Minor.Injuries', 'Total.Uninjured']
for col in numerical_columns:
    df_cleaned[col] = df_cleaned[col].fillna(0)  # Replacing Null values with 0
    print(df_cleaned[col])
```

```python
categorical_columns = df_cleaned.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

```python
categorical_columns = ['Event.Id', 'Investigation.Type', 'Accident.Number', 'Event.Date',
       'Location', 'Country',  'Aircraft.damage', 'Make',
       'Model', 'Amateur.Built', 'Engine.Type', 'Purpose.of.flight',
       'Weather.Condition','Report.Status',
       'Publication.Date', 'Injury.Severity','Registration.Number']
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].fillna('Unknown')  
```

Objective 1: Identify aircraft models with a lower likelihood of severe accidents or operational failures

```python
Accidents = df1[df1['Investigation_Type'] == 'Accident']
Make_Accidents = Accidents.groupby('Make').size().sort_values(ascending=False)
Model_Accidents = Accidents.groupby('Model').size().sort_values(ascending=False)
# We now print the top 15 Aircrafts with the most accident occurences both the model and the make
print(f'The top 15 Aircraft Makes with the most accident occurences:\n{Make_Accidents.head(15)}\n\n\n')
print(f'Top 15 Aircraft Models with the most accident occurences:\n{Model_Accidents.head(15)}\n\n\n')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# (Aircraft Models)
axes[0].bar(Model_Accidents.head(15).index, Model_Accidents.head(15).values, color='purple', width=0.6)
axes[0].set_xlabel('Aircraft Model')
axes[0].set_ylabel('Number of Accidents')
axes[0].set_title('aircraft models with a lower likelihood of severe accidents')
axes[0].tick_params(axis='x', rotation=45)

# (Aircraft Makes)
axes[1].bar(Make_Accidents.head(15).index, Make_Accidents.head(15).values, color='blue', width=0.6)
axes[1].set_xlabel('Aircraft Make')
axes[1].set_ylabel('Number of Accidents')
axes[1].set_title('aircraft makes with a lower likelihood of severe accidents')
axes[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

```

Objective 2: Determine locations that pose higher risks for aircraft operations

```python
location_accidents = df1.groupby('Location').size().sort_values(ascending=False).head(15)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=location_accidents.values, y=location_accidents.index, palette="Reds_r", ax=ax)
ax.set_xlabel("Number of Accidents")
ax.set_ylabel("Location")
ax.set_title("Locations that poses risks for aircraft operations")
ax.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

Objective 3:  Assess the optimal number of engines required for stable and safe flights

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df1['Engine.Type'].dropna(), bins=10, color="skyblue", edgecolor="black", alpha=0.7)
ax.set_xlabel("Engine Type")
ax.set_ylabel("Frequency of Accidents")
ax.set_title("Histogram of Accidents by Engine Type")
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.xticks(rotation=45)  
plt.show()

```

CONCLUSIONS
 
 From the objectives above i can conclude that:
 
Safer Aircraft Models: Single-engine planes (e.g., Cessna 172, Piper PA-28) are frequently involved in accidents but with lower fatality rates. Beechcraft models and helicopters also show relatively safer outcomes in certain conditions.

High-Risk Locations: Mountainous regions (e.g., Alaska, Colorado) and urban areas (e.g., Los Angeles, New York) see more accidents due to weather, terrain, and traffic complexity. Rural areas also pose risks due to limited infrastructure.

Optimal Engine Count: Single-engine aircraft are safe in good conditions but riskier in emergencies. Twin-engine planes offer redundancy but face more severe conditions. Turbojet/turboprop aircraft are highly reliable but have higher fatality rates in accidents.

RECOMMENDATIONS

1. Single-Engine Aircraft: Improve pilot training, adopt modern safety technologies, and enhance weather awareness to reduce risks.

2. High-Risk Locations: Strengthen weather forecasting, air traffic management, and emergency infrastructure in mountainous, urban, and rural areas.

3. Engine Optimization: Encourage multi-engine aircraft for high-risk flights, integrate advanced avionics in single-engine planes, and enforce stricter safety guidelines for turbojet/turboprop operations
