```python
sorted_df = missing_df.sort_values(ascending=False)
total_rows = df.shape[0]
missing_df_percentage = (missing_df / total_rows) * 100
missing_df_percentage = missing_df_percentage.sort_values(ascending=False).round(0)
```

(ii) Objective 1:  Uncover the Aircraft Models That are Safest for Air Travel

```python
Accidents = df1[df1['Investigation_Type'] == 'Accident']
Make_Accidents = Accidents.groupby('Make').size().sort_values(ascending=False)
Model_Accidents = Accidents.groupby('Model').size().sort_values(ascending=False)
print(f'The top 15 Aircraft Makes with the most accident occurences:\n{Make_Accidents.head(15)}\n\n\n')
print(f'Top 15 Aircraft Models with the most accident occurences:\n{Model_Accidents.head(15)}\n\n\n')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# (Aircraft Models)
axes[0].bar(Model_Accidents.head(15).index, Model_Accidents.head(15).values, color='purple', width=0.6)
axes[0].set_xlabel('Aircraft Model')
axes[0].set_ylabel('Number of Accidents')
axes[0].set_title('aircraft models with a lower likelihood of severe accidents')
axes[0].tick_params(axis='x', rotation=45)

# (Aircraft Makes)
axes[1].bar(Make_Accidents.head(15).index, Make_Accidents.head(15).values, color='blue', width=0.6)
axes[1].set_xlabel('Aircraft Make')
axes[1].set_ylabel('Number of Accidents')
axes[1].set_title('aircraft makes with a lower likelihood of severe accidents')
axes[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

```

(iii) Objective 2: Determine locations that pose higher risks for aircraft operations

```python
location_accidents = df1.groupby('Location').size().sort_values(ascending=False).head(15)

fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=location_accidents.values, y=location_accidents.index, palette="Greens_r", ax=ax)  

ax.set_xlabel("Number of Accidents")
ax.set_ylabel("Location")
ax.set_title("Top 15 High-Risk Locations for Aircraft Accidents")
ax.grid(axis="x", linestyle="--", alpha=0.7)

plt.show()
```

(iv) Objective 3: Identify how many engines Does a Plane Really Need

```python
fatal_rate = df1.groupby('Number_of_Engines')['Total_Fatal_Injuries'].mean()
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=fatal_rate.index, y=fatal_rate.values, palette="Reds_r", ax=ax)
ax.set_xlabel("Number of Engines")
ax.set_ylabel("Average Fatalities per Accident")
ax.set_title("Fatal Accident Rate vs. Engine Count")
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()



```
